```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#print(sns.get_dataset_names())

sns.set_theme("notebook")
iris = sns.load_dataset("iris")
iris.head(10)

sns.pairplot(iris, hue="species")
plt.show()


```

```python
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.show()
```

```python
sns.distplot(iris["sepal_length"])
plt.show()
```

```python
sns.histplot(iris["sepal_length"], bins = 20)
plt.show()
```

```python
sns.boxplot(data=iris, x="species", y="sepal_width", hue= "species")
plt.show()
```

```python
penguins = sns.load_dataset("penguins")

plt.title("Penguin Body Mass based on Gender and Species")
sns.barplot(data=penguins, x="species", y="body_mass_g", hue="sex")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pokemon= pd.read_csv("pokRdex_comma.csv")

pokemon.head()
sns.violinplot(data=pokemon, x="type_1",  y="special_attack",hue="type_1")
plt.show()

```

```python
pokemon.head()
```

```python
import matplotlib.pyplot as plt

plt.title("Special Attack based on Pokemon Type")
plt.xlabel("Type")
plt.ylabel("Special Attack")

plt.barh(pokemon["type_1"], pokemon["special_attack"], height=0.5)
plt.show()
```

```python
type_counts = pokemon["type_1"].value_counts()

ype_counts = pokemon["type_1"].value_counts()  # top 5 for clarity

# Donut chart
plt.pie(
    type_counts.values,
    labels=type_counts.index,
    autopct="%.1f%%",
    startangle=140,
    wedgeprops=dict(width=0.4),
    colors=plt.cm.Set3.colors
)

plt.title(" Pokemon Types (Donut Chart)")
plt.axis("equal")
plt.text(0, 0, f"Total\n{type_counts.sum()}", ha='center', va='center', fontsize=12)
plt.show()
```
